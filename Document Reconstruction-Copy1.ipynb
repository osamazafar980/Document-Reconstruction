```python
def match(word,deep):
    word=word.lower()
    file1 = open('words.txt', 'r') 
    lines = file1.readlines() 
    for line in lines:
        if(deep):
            if(line.startswith(word[:-1])):
                return True
        else:
            if(line.strip()==word.lower()):
                return True
            
    return False
def matchEnds(word):
    word=word.lower()
    file1 = open('words.txt', 'r') 
    lines = file1.readlines() 
    for line in lines:
        if(len(word)>4):
            if(line.strip().endswith(word)):
                return True
        else:
            if(line.strip()==word):
                return True
            
    return False

def removePunc(in_string):
  out_string= in_string.replace("\n"," ")
  for i in out_string:
    if not i.isalpha():
      out_string=out_string.replace(i," ")
  return out_string.lower()

def clean(word):
    lines=word.split(" ")
    line=[]
    status=False
    for i in lines:
        if(i!=""):
            status=True
            line.append(i)
    return line,status

def display(lines):
    for i in lines:
        print(i)

def scaner(img):
    img1 = img.astype('uint8')
    cv.cvtColor(img1,cv.COLOR_BGR2RGB)
    words = tess.image_to_data(img1)
    word=''
    lines=[]
    for x,w in enumerate(words.splitlines()):
        w=w.split()
        if(str((w[-1]))!="-1"):
            word=word+" "+w[-1]
        else:
            line,status= clean(removePunc(word))
            if(status):
                lines.append(line)

            word=''
    return lines
def getLeftIndexes(parts,deep):
    left=[]
    for i in range(len(parts)):
        part=parts[i]
        lines=scaner(part)
        status=False
        for line in lines:
            status=match(line[0],deep)
            if(not(status)):
                break;
        if(status):
            left.append(i)
    return left
def getRightIndexes(parts):
    right=[]
    for i in range(len(parts)):
        part=parts[i]
        lines=scaner(part)
        status=False
        for line in lines:
            status=matchEnds(line[-1])
            if(not(status)):
                right.append(0)
                break;
        if(status):
            right.append(1)
    return right

def getRightScores(parts):
    right=[]
    for i in range(len(parts)):
        part=cv.cvtColor(parts[i],cv.COLOR_BGR2GRAY)
        score=0
        BWImage=part[:,(part.shape[1]//4)*3:]
        for i in range(len(BWImage)):
            for j in range(len(BWImage[i])):
                if(BWImage[i][j]>200):
                    score+=1
        right.append(score)
    return right

def stripEdge(data):
    lines=[]
    for line in data:
        lines.append(line[:-1])
    return lines
def score(data,deep):
    score=0
    for line in data:
        if len(line)>0:
            for word in line:
                if(not(match(word,deep))):
                    score=score+1
    return score

```

```python
parts=[]
names=[]
for i in range(1,36):
    img=cv.imread('img/P'+str(i)+'.png')
    x=img.shape[1]//4
    img1=img[:,:x,:]
    img2=img[:,x:x*2,:]
    img3=img[:,x*2:x*3,:]
    img4=img[:,x*3:,:]
    parts.append(img1)
    parts.append(img2)
    parts.append(img3)
    parts.append(img4)
    names.append('P'+str(len(parts)-3))
    names.append('P'+str(len(parts)-2))
    names.append('P'+str(len(parts)-1))
    names.append('P'+str(len(parts)))
for i in range(len(parts)//4):
    v1=randint(0, len(parts))
    v2=randint(0, len(parts))
    temp=parts[v1]
    parts[v1]=parts[v2]
    parts[v2]=temp
    temp=names[v1]
    names[v1]=names[v2]
    names[v2]=temp

print(names)
partscopy=copy.deepcopy(parts)
namescopy=copy.deepcopy(names)
```

```python
parts=partscopy
names=namescopy
len(namescopy)
```

```python
print("Enter a Number to Choose:")
print("1. shallow reconstruction")
print("2. deep reconstruction")
c=int(input("Choice :"))
deep=True
if(c==1):
    deep=False
```

```python
leftIndexes=getLeftIndexes(parts,deep)
rightIndexes=getRightIndexes(parts)
print(leftIndexes)
print(rightIndexes)
print(names)
leftParts=[]
i=0
for ind in leftIndexes:
    leftParts.append(parts[ind-i])
    del parts[ind-i]
    del names[ind-i]
    del rightIndexes[ind-i]
    i=i+1
print(rightIndexes)
print(names)
```

```python
image=leftParts[0]
j=0
for leftPart in leftParts:
    image=leftPart
    for i in range(len(parts)):
        matchScore=[]
        inds=[]
        counter=0
        for part in parts:
            if (image.shape[0]==part.shape[0]):
                crossover=np.concatenate((image,part),axis=1)
                data=scaner(crossover)
                if(not(deep)):
                    data=stripEdge(data)
                matchScore.append(score(data,deep))
                inds.append(counter)
            counter=counter+1
            print(str(matchScore)+"\t"+str(inds))
        if(len(inds)!=0):
            print("merging "+str(names[inds[matchScore.index(min(matchScore))]]))
            image=np.concatenate((image,parts[inds[matchScore.index(min(matchScore))]]),axis=1)
            if(rightIndexes[inds[matchScore.index(min(matchScore))]]==1):
                del parts[inds[matchScore.index(min(matchScore))]]
                del names[inds[matchScore.index(min(matchScore))]]
                del rightIndexes[inds[matchScore.index(min(matchScore))]]
                print(names)
                print(rightIndexes)
                break
            del parts[inds[matchScore.index(min(matchScore))]]
            del names[inds[matchScore.index(min(matchScore))]]
            del rightIndexes[inds[matchScore.index(min(matchScore))]]
        print(names)
        print(rightIndexes)
                    
    cv.imwrite('reconstruced'+str(j)+'.png', image)
    j=j+1
    print('reconstruced'+str(j-1)+'.png')

    
```
